In this exercise, we will practice **variables**, **data types**, and **if/elif/else statements** using a fun theme: a **Smoothie Shop**. Imagine you are helping a small juice bar build their very first ordering system.

## ü•§ Step 1: Variables and Data Types
- Create variables to store the following information:
  - The name of a smoothie (string)
  - The price of a small smoothie (float)
  - The price of a large smoothie (float)
  - The daily special (boolean: `True` if available, `False` if not)

```
smoothie_name = ""
small_price = 0.0
large_price = 0.0
daily_special = False
```

## ü•ù Step 2: Getting Input from the User
- Ask the customer for their name.
- Ask them which size they want: "small" or "large".

```
customer_name = input("Enter your name: ")
size = input("Do you want a small or large smoothie? ")
```

## üçì Step 3: Calculate the Price
- Use an `if/elif/else` statement to decide the price based on the size chosen.
- If the customer enters something other than "small" or "large", set the price to 0 and display an error message.

```
if size == "small":
    price = small_price
elif size == "large":
    price = large_price
else:
    price = 0
    print("Invalid size!")
```

## üçç Step 4: Add a Discount for Specials
- If `daily_special` is `True`, give the customer a 20% discount.

```
if daily_special:
    price = price * 0.8
```

## üçä Step 5: Display the Final Bill
- Print a friendly message to the customer showing:
  - Their name
  - The smoothie they ordered
  - The final price

```
print("Hello,", customer_name)
print("You ordered a", size, smoothie_name)
print("Your total is: $", price)
```
