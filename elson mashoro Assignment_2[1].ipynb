```python
# TODO: Write your solution here

name = input("Enter a name: ")
match name:
    case "Harry" | "Hermione" | "Ron":
        print("Gryffindor")
    case "Draco":
        print("Slytherin")
    case _:
        print("Unknown")

# Use match-case here (Python 3.10+)
```
